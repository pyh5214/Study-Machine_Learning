## 전처리 - 피처 엔지니어링

### 나이(내나이와 상대방의 나이)

```python
def age_gap(x):
    if x['age'] == -99: #결측처리가 되어있는 경우
        return -99
    elif x['age_o'] == 99:
        return -99
    elif x['gender'] == 'female':
        return x['age_o'] - x['age']
    else:
        return x['age'] - x['age_o']
```

- 남녀 중 누구 하나라도 결측일 경우 결츠값 반환
- 남자가 연상이면 +값, 여자가 연상이면 -값

```python
data['age_gap'] = data.apply(age_gap, axis = 1)
```

### 인종
- 본인과 상대방이 같은 인종이면 1, 그렇지 않으면 -1, 결측값 -99도 존재

```python
def same_race(x):
    if x['race'] == -99:
        return -99
    elif x['race_o'] == -99:
        return -99
    elif x['race'] == x['race_o']:
        return 1
    else:
        return -1
```

```python
data['same_race'] = data.apply(same_race, axis = 1)
```

```python
def same_race_point(x):
    if x['same_race'] == -99:
        return -99
    else:
        return x['same_race'] * x['importance_same_race']
```

```python
data['same_race_point'] = data.apply(same_race_point, axis = 1)
```

### 평가항목별 점수 환산
- 평가점수 * 중요도

```python
def rating(data, importance, score):
    if data[importance] == -99:
        return -99
    elif data[score] == -99:
        return -99
    else:
        return data[importance] * data[score]
```

- pref_o_xxx : 상대방이 xxx 항목을 얼마나 중요하게 생각하는지에 대한 응답
- xxx_o : 상대방이 본인에 대한 xxx 항목을 평가한 값
- xxx_importance : xxx 항목에 대해 본인이 얼마나 중요하게 생각하는지에 대한 응답
- xxx_partner : 본인이 상대방에 대한 xxx 항목을 평가한 값

```python
# 상대방의 중요도
```

```python
# 'pref_o_attractive','pref_o_sincere', 'pref_o_intelligence', 'pref_o_funny','pref_o_ambitious', 'pref_o_shared_interests'
```

```python
# 상대방의 중요도
partner_imp = data.columns[8:14]
# 상대방이 본인에 대한 평가
partner_rate_me = data.columns[14:20]
# 본인의 중요도
my_imp = data.columns[20:26]
# 상대방에 대한 본인의 평가
my_rate_partner = data.columns[26:32]
```

```python
new_partner = ['attractive_p', 'sincere_p', 'intelligence_p', 'funny_p', 'ambition_p', 'shared_interests_p']
new_me = ['attractive_m', 'sincere_m', 'intelligence_m', 'funny_m', 'ambition_m', 'shared_interests_m']
```

```python
# 상대방 관련 항목 점수
for i, j, k in zip(new_partner, partner_imp, partner_rate_me):
    data[i] = data.apply(lambda x : rating(x, j, k), axis = 1)
```

```python
# 본인 관련 항목 점수
for i, j, k in zip(new_me, my_imp, my_rate_partner):
    data[i] = data.apply(lambda x : rating(x, j, k), axis = 1)
```

# 모델링과 평가

## 훈련세트와 테스트세트 분리

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop('match', axis = 1), 
                                                    data['match'], test_size = 0.2, random_state = 100)
```

## 모델링과 평가

```python
import xgboost as xgb

model = xgb.XGBClassifier(n_estimators = 500, max_depth = 5, random_state = 100)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_score(y_test, pred)
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```
